```python
fig, ax = plt.subplots()
ax.plot(t, xa, 'r-', label=f'$x(t)$')
ax.plot(t, ya, 'b-', label=f'$y(t)$')
ax.set_xlabel('$t$')
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 10)
ax.legend(frameon=False, loc='upper right')
plt.show()
```
